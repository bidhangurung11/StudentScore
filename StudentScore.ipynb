```python
#drop unnamed column
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("GENDER DISTRIBUTION")
plt.show()
```

```python
#GROUP BY PARENTS EDUCATION
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb , annot = True)
plt.title("RELATIONSHIP BETWEEN PARENTS EDU AND CHILDREN'S SCORE")
plt.show()
```

From the above chart there is impact of parents education on their childern's marks.

##DISTRIBUTION OF ETHNIC GROUPS


```python
data = {'EthnicGroup': ['group A', 'group B', 'group C', 'group A', 'group B', 'group D', 'group A', 'group C', 'group E']}
df = pd.DataFrame(data)

# Count occurrences for each group
group_counts = df['EthnicGroup'].value_counts()

# Extract the counts for specific groups
groups = ["group A", "group B", "group C", "group D", "group E"]
mlist = [group_counts.get(group, 0) for group in groups]  # Use `.get()` to handle missing groups

# Plot the pie chart
plt.pie(mlist, labels=groups, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
